## Data visualization of Boston Housing dataset

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_boston
boston_dataset = load_boston()
```

```python
print(boston_dataset.keys())
```

data = contains information about various houses

target = prices of the house

feature_names = names of the features

DESCR = describes the dataset

```python
boston_dataset.DESCR
```

```python
boston = pd.DataFrame(boston_dataset.data, columns=boston_dataset.feature_names)
boston.head()
```

```python
# Target value MEDV is missing from the data, so we create a new column
boston['MEDV'] = boston_dataset.target
```

```python
boston.isnull().sum()
```

```python
# Plot the distribution of the target variable MEDV
sns.set(rc={'figure.figsize': (11.7, 8.27)})
sns.displot(boston['MEDV'], bins=30, kde=True) # normal distribution kde
plt.show()
```

```python
# Plot the distribution of the target variable MEDV
sns.set(rc={'figure.figsize': (11.7, 8.27)}) # Gaussian kernel density estimate
sns.distplot(boston['MEDV'], bins=30)
plt.show()
```

```python
correlation_matrix = boston.corr().round(2)
# annot = True to print the values inside the squares
sns.heatmap(data=correlation_matrix, annot=True)
plt.show()
```
