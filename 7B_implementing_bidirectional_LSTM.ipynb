## Bidirectional LSTM

```python
from random import random
from numpy import array
from numpy import cumsum
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dense
from keras.layers import TimeDistributed
```

```python
from keras.layers import Bidirectional
```

```python
def get_sequence(n_timesteps):
    # Create a sequence of random numbers in [0, 1]
    x1 = array([random() for _ in range(n_timesteps)])

    # Calculate cut-off  value to change class values
    limit = n_timesteps/4.0

    # Determine the class outcome for each item in cumuative sequence
    y = array([0 if i<limit else 1 for i in cumsum(x1)])

    # Reshape input and output data to be suitable for LSTMs
    x1 = x1.reshape(1, n_timesteps, 1)
    y = y.reshape(1, n_timesteps, 1)

    return x1, y
```

```python
# Define problem properties
n_timesteps = 10
```

```python
# Define LSTM
model = Sequential()
model.add(Bidirectional(LSTM(20, input_shape=(n_timesteps,1), return_sequences=True)))
model.add(TimeDistributed(Dense(1, activation='sigmoid')))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['acc'])
```

```python
# Train LSTM
for epoch in range(1000):

    # Generate a new random sequence
    x1, y = get_sequence(n_timesteps)

    # Fit model for one epoch on this sequence
    model.fit(x1, y, epochs=1, batch_size=1, verbose=2)
```

```python
# Evaluate LSTM

x1, y = get_sequence(n_timesteps)
yhat = model.predict_classes(x1, verbose=0)
print(x1)
print(yhat)
```
